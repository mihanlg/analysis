## Регуляризация

```python
%matplotlib inline
```

Под регуляризацией обычно подразумевается добавление некоторых дополнительных ограничений к искомому решению с целью получения решения, устойчивого к переобучению. В случае с линейной регрессией, полученное решение будет также обладать полезными статистическими свойствами.

Для иллюстрации сгенерируем набор данных и попробуем его аппроксимировать полиномами различной степени.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

from sklearn.cross_validation import train_test_split

try:
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.pipeline import make_pipeline
except ImportError:
    # use backports for sklearn 1.4
    from sklearn_backports import PolynomialFeatures
    from sklearn_backports import make_pipeline

# ignore DeprecateWarnings by sklearn
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

np.random.seed(9)

def f(x):
    return np.sin(2 * np.pi * x)

x_plot = np.linspace(0, 1, 100)

n_samples = 100
X = np.random.uniform(0, 1, size=n_samples)[:, np.newaxis]
y = f(X) + np.random.normal(scale=0.3, size=n_samples)[:, np.newaxis]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)

ax = plt.gca()
ax.plot(x_plot, f(x_plot), color='green')
ax.scatter(X_train, y_train, s=10)
ax.set_ylim((-2, 2))
ax.set_xlim((0, 1))
ax.set_ylabel('y')
ax.set_xlabel('x')
```

```python
def plot_approximation(est, ax, label=None):
    ax.plot(x_plot, f(x_plot), color='green')
    ax.scatter(X_train, y_train, s=10)
    ax.plot(x_plot, est.predict(x_plot[:, np.newaxis]), color='red', label=label)
    ax.set_ylim((-2, 2))
    ax.set_xlim((0, 1))
    ax.set_ylabel('y')
    ax.set_xlabel('x')
    ax.legend(loc='upper right')  #, fontsize='small')

fig, axes = plt.subplots(2, 2, figsize=(8, 5))
for ax, degree in zip(axes.ravel(), [0, 1, 3, 9]):
    est = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    est.fit(X_train, y_train)
    plot_approximation(est, ax, label='degree=%d' % degree)
    
plt.tight_layout()
```

Происходящее понятно: мы минимизируем наблюдаемую ошибку и "идеальный" с этой точки зрения алгоритм - тот, который просто "запомнил" точки, на которых он был обучен.

**Как минимум одно решение этой проблемы вам известно.**

```python
from sklearn.metrics import mean_squared_error

train_error = np.empty(10)
test_error = np.empty(10)
for degree in range(10):
    est = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    est.fit(X_train, y_train)
    train_error[degree] = mean_squared_error(y_train, est.predict(X_train))
    test_error[degree] = mean_squared_error(y_test, est.predict(X_test))

plt.plot(np.arange(10), train_error, color='green', label='train')
plt.plot(np.arange(10), test_error, color='red', label='test')
plt.ylim((0.0, 1e0))
plt.ylabel('log(mean squared error)')
plt.xlabel('degree')
plt.legend(loc='lower left')
```

### Ридж регрессия

Первый из методов, которые мы рассмотрим - ридж регрессия, известная так же как гребневая регрессия, частный случай регуляризации Тихонова.

Добавим к целевой функции ограничение в норме $L_2$:

$$ RSS(\lambda) = (y-X\beta)^T(y-X\beta) + \lambda\beta^T \beta.$$

тогда решение будет в виде:

$$ \hat{\beta}_{ridge} = (X^TX + \lambda I)^{-1}X^Ty.$$

Из-за нового слагаемого $\lambda I$ она и получила название "гребневая". Можно показать, что полученные решения являются решениями следующей задачи минимизации:

$$\hat{\beta}_{ridge} = arg\,min \sum_{i=1}^{n} y_i - \beta x_i, $$
при условии, что 
$$ \sum_{j=1}^{p} \beta_j^2 \leq t.$$

```python
fig, ax_rows = plt.subplots(4, 2, figsize=(8, 10))

def plot_coefficients(est, ax, label=None, yscale='log'):
    coef = est.steps[-1][1].coef_.ravel()
    if yscale == 'log':
        ax.semilogy(np.abs(coef), marker='o', label=label)
        ax.set_ylim((1e-1, 1e8))
    else:
        ax.plot(np.abs(coef), marker='o', label=label)
    ax.set_ylabel('abs(coefficient)')
    ax.set_xlabel('coefficients')
    ax.set_xlim((1, 9))

degree = 9
alphas = [1e-8, 1e-3, 1e-2, 1e-1]
for alpha, ax_row in zip(alphas, ax_rows):
    ax_left, ax_right = ax_row
    est = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=alpha))
    est.fit(X_train, y_train)
    plot_approximation(est, ax_left, label='alpha=%r' % alpha)
    plot_coefficients(est, ax_right, label='Ridge(alpha=%r) coefficients' % alpha)

plt.tight_layout()
```

Что нам это дает?


**С точки зрения контроля над переобучением**: мы "не разрешаем" оценкам модели принимать слишком большие или слишком маленькие значения, тем самым контролируя сложность решения;

**С точки зрения качества модели**: если объясняющие переменные коррелируют между собой (т.е. наблюдается мультиколлинеарность), то оценки параметров будут численно нестабильными, а матрица $X^TX$ близка к сингулярной. Введение дполнительных параметров делает оценки смещенными, но состоятельными.


### LASSO

Как и в случае регуляризации Тихонова, "оштрафуем" решение на величину значений параметров, только не в $L_2$ норме, а в $L_1$:

$$\hat{\beta}_{ridge} = arg\,min \sum_{i=1}^{n} y_i - \beta x_i, $$
при условии, что 
$$ \sum_{j=1}^{p} \lvert\beta_j\rvert \leq t.$$

Из формы решения видно, что LASSO зануляет некоторые коэффициенты. 

```python
from sklearn.linear_model import Lasso

fig, ax_rows = plt.subplots(4, 2, figsize=(8, 10))

degree = 9

alphas = [1e-8, 1e-3, 1e-2, 1e-1]
for alpha, ax_row in zip(alphas, ax_rows):
    ax_left, ax_right = ax_row
    est = make_pipeline(PolynomialFeatures(degree), Lasso(alpha=alpha))
    est.fit(X_train, y_train)
    plot_approximation(est, ax_left, label='alpha=%r' % alpha)
    plot_coefficients(est, ax_right, label='Lasso(alpha=%r) coefficients' % alpha, yscale=None)
    
plt.tight_layout()
```

Преимущество метода в том, что он выдает sparse решение, которое может использоваться для отбора наиболее информативных признаков.

## Elastic net


Как всегда, наш "любимый" вариант - линейная комбинация двух вышеописанных методов. Наряду с отбором признаков, позволяет контроллировать величину необнуленных коэффициентов.

```python
from sklearn.linear_model import ElasticNet

fig, ax_rows = plt.subplots(4, 2, figsize=(8, 10))

degree = 9

alphas = [1e-8, 1e-3, 1e-2, 1e-1]
for alpha, ax_row in zip(alphas, ax_rows):
    ax_left, ax_right = ax_row
    est = make_pipeline(PolynomialFeatures(degree), ElasticNet(alpha=alpha, l1_ratio=0.5))
    est.fit(X_train, y_train)
    plot_approximation(est, ax_left, label='alpha=%r' % alpha)
    plot_coefficients(est, ax_right, label='EN(alpha=%r) coefficients' % alpha, yscale=None)
    
plt.tight_layout()
```
